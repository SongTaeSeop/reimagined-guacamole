```python
import matplotlib.pyplot as plt

# matplotlib은 데이터를 차트나 플롯으로 시각화

plt.title("test")
plt.plot([1, 2, 3, 4], [2, 4, 8, 6])
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title("add X, Y Label")
plt.plot([1, 2, 3, 4], [2, 4, 8, 6])
plt.xlabel("Hours")
plt.ylabel("Score")
plt.show()
```

```python
plt.title("add legend")
plt.plot([1, 2, 3, 4], [2, 4, 8, 6])
plt.plot([1.5, 2.5, 3.5, 4.5], [3, 5, 8, 10])
plt.xlabel("Hours")
plt.ylabel("Score")
plt.legend(["A Class", "B Class"])
plt.show()
```
